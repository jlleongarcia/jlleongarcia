```python
import re
import matplotlib.pyplot as plt
from geopy.geocoders import Nominatim
from mpl_toolkits.basemap import Basemap

# Define the file to parse
file_path = "output.txt"

# Read the file
with open(file_path, "r") as file:
    data = file.read()

# Extract GPS Latitude and Longitude values using regex
latitude_matches = re.findall(r"GPS Latitude\s+:\s+([\d.]+) deg ([\d.]+)' ([\d.]+)\" ([NS])", data)
longitude_matches = re.findall(r"GPS Longitude\s+:\s+([\d.]+) deg ([\d.]+)' ([\d.]+)\" ([EW])", data)

# Function to convert to decimal degrees
def dms_to_decimal(degrees, minutes, seconds, direction):
    decimal = float(degrees) + float(minutes)/60 + float(seconds)/3600
    if direction in ["S", "W"]:
        decimal = -decimal
    return decimal

# Convert matches to decimal degrees
latitudes = [dms_to_decimal(*match) for match in latitude_matches]
longitudes = [dms_to_decimal(*match) for match in longitude_matches]

# Pair the latitudes and longitudes
coordinates = list(zip(latitudes, longitudes))

# Take the first 10 pairs
first_10_coordinates = coordinates[:10]

# Print the extracted coordinates
print("First 10 GPS Coordinates:")
for i, (lat, lon) in enumerate(first_10_coordinates, start=1):
    print(f"{i}: Latitude: {lat}, Longitude: {lon}")

# Plotting the coordinates on a map
plt.figure(figsize=(10, 7))

# Setup Basemap
m = Basemap(projection='merc', resolution='i',
            llcrnrlat=min(latitudes) - 1, urcrnrlat=max(latitudes) + 1,
            llcrnrlon=min(longitudes) - 1, urcrnrlon=max(longitudes) + 1)

m.drawcoastlines()
m.drawcountries()
m.drawmapboundary(fill_color='aqua')
m.fillcontinents(color='lightgreen', lake_color='aqua')

# Convert coordinates to map projection
x, y = m([lon for lat, lon in first_10_coordinates], [lat for lat, lon in first_10_coordinates])

# Plot points
m.scatter(x, y, color='red', marker='o', zorder=5, label="GPS Points")
plt.title("First 10 GPS Coordinates")
plt.legend()
plt.show()
```
